```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
```

## Maldición de dimensionalidad y polynomial features
#### En estadistica la maldicion de dimensionalidad es también conocida como efecto Hughes, se refiere a los diversos fenómenos que surgen al analizar y organizar datos de espacios de múltiples dimensiones, en nuestro caso los aplicaremos a datasets que simulan multiples variables. En el caso de polynomial features son todas las combinaciones polinomiales de las features con grado menor o igual que el grado especificado en la funcion, por ejemplo si hay un input de dos dimensiones la forma de grado dos polinomial para el conjunto a,b seria el siguiente [1, a, b, a^2, ab, b^2].


#### Random values de 1 feature y 10 observaciones  d0=10,  d1=1, LAS CUALES SE CONVIERTEN EN 3 FEATURES

```python
var1 = np.random.rand(10,1)
print('Valores', var1)
print('')
print('Polinomiales')
var_pol1 = PolynomialFeatures(3)
pd.DataFrame(var_pol1.fit_transform(var1))
```

#### Random values de 2 features y 10 observaciones  d0=10,  d1=2, LAS CUALES SE CONVIERTEN EN 9 FEATURES

```python
var2 = np.random.rand(10,2)
print('Valores', var1)
print('')
print('Polinomiales')
var_pol2 = PolynomialFeatures(3)
pd.DataFrame(var_pol1.fit_transform(var2))
```

#### Random values de 2 features y 10 observaciones  d0=10,  d1=3, LAS CUALES SE CONVIERTEN EN 19 FEATURES

```python
var3 = np.random.rand(10,3)
print('Valores', var3)
print('')
print('Polinomiales')
var_pol3 = PolynomialFeatures(3)
pd.DataFrame(var_pol1.fit_transform(var3))
```

### CONCLUSIONES
Cuando se aumenta el grado del polinomio se incrementa exponencialmente las features en el dataset
